```python
# Calculando o RSL de 10 periodos

p = 10

df_mapa["MM_indice"] = df_mapa["^BVSP"].rolling(p).mean()
df_mapa["MM_ativo1"] = df_mapa["PETR4.SA"].rolling(p).mean()
df_mapa["MM_ativo2"] = df_mapa["FESA4.SA"].rolling(p).mean()
df_mapa["MM_ativo3"] = df_mapa["JBSS3.SA"].rolling(p).mean()
df_mapa["MM_ativo4"] = df_mapa["VALE3.SA"].rolling(p).mean()
df_mapa["MM_ativo5"] = df_mapa["GGBR4.SA"].rolling(p).mean()
```

```python
df_mapa["RSL_indice"] = (df_mapa["^BVSP"]/df_mapa["MM_indice"] - 1)*100
df_mapa["RSL_ativo1"] = (df_mapa["PETR4.SA"]/df_mapa["MM_ativo1"] - 1)*100
df_mapa["RSL_ativo2"] = (df_mapa["FESA4.SA"]/df_mapa["MM_ativo2"] - 1)*100
df_mapa["RSL_ativo3"] = (df_mapa["JBSS3.SA"]/df_mapa["MM_ativo3"] - 1)*100
df_mapa["RSL_ativo4"] = (df_mapa["VALE3.SA"]/df_mapa["MM_ativo4"] - 1)*100
df_mapa["RSL_ativo5"] = (df_mapa["GGBR4.SA"]/df_mapa["MM_ativo5"] - 1)*100
```

```python
# Calculando o RSI (IFR) de 2 e 14 periodos


indice_rsi2 = ta.momentum.RSIIndicator(close = df_mapa["^BVSP"], window = 2)
indice_rsi14 = ta.momentum.RSIIndicator(close = df_mapa["^BVSP"], window = 14)
ativo1_rsi2 = ta.momentum.RSIIndicator(close = df_mapa["PETR4.SA"], window = 2)
ativo1_rsi14 = ta.momentum.RSIIndicator(close = df_mapa["PETR4.SA"], window = 14)
ativo2_rsi2 = ta.momentum.RSIIndicator(close = df_mapa["FESA4.SA"], window = 2)
ativo2_rsi14 = ta.momentum.RSIIndicator(close = df_mapa["FESA4.SA"], window = 14)
ativo3_rsi2 = ta.momentum.RSIIndicator(close = df_mapa["JBSS3.SA"], window = 2)
ativo3_rsi14 = ta.momentum.RSIIndicator(close = df_mapa["JBSS3.SA"], window = 14)
ativo4_rsi2 = ta.momentum.RSIIndicator(close = df_mapa["VALE3.SA"], window = 2)
ativo4_rsi14 = ta.momentum.RSIIndicator(close = df_mapa["VALE3.SA"], window = 14)
ativo5_rsi2 = ta.momentum.RSIIndicator(close = df_mapa["GGBR4.SA"], window = 2)
ativo5_rsi14 = ta.momentum.RSIIndicator(close = df_mapa["GGBR4.SA"], window = 14)



df_mapa["RSI2_BVSP"] = indice_rsi2.rsi()
df_mapa["RSI14_BVSP"] = indice_rsi14.rsi()

df_mapa["RSI2_PETR"] = ativo1_rsi2.rsi()
df_mapa["RSI14_PETR"] = ativo1_rsi14.rsi()

#arrumar a partir daqui

df_mapa["RSI2_FESA"] = itub_rsi2.rsi()
df_mapa["RSI14_FESA"] = itub_rsi14.rsi()

df_mapa["RSI2_VALE"] = vale_rsi2.rsi()
df_mapa["RSI14_VALE"] = vale_rsi14.rsi()

df_mapa["RSI2_JBSS"] = bbdc_rsi2.rsi()
df_mapa["RSI14_JBSS"] = bbdc_rsi14.rsi()

df_mapa["RSI2_GGBR"] = bbas_rsi2.rsi()
df_mapa["RSI14_GGBR"] = bbas_rsi14.rsi()
```
